```python
import nltk
from nltk.corpus import cmudict
from nltk.tokenize import TweetTokenizer
```

```python
folderPath = "C:\\Users\\katar\\Documents\\2017\\computerMusic\\projects\\rhymesToMusic\\"
fileName = "sourceText\\raven.txt"
phonesFileName="word_pronounciation_dictionary\\cmudict-master\\cmudict-master\\cmudict.phones"
```

```python
d = cmudict.dict()
```

```python
s = "'Tis some visitor, I muttered, tapping at the chamber door, Only this, and nothing more."
inputFile = folderPath + fileName
with open(file=inputFile,mode='r',encoding='utf-8') as inputText:
    read_data = inputText.read()
```

```python
tknzr = TweetTokenizer()
```

```python
tlist = tknzr.tokenize(read_data.lower())
```

```python
len(tlist)
```

```python
for t in range(0,len(tlist)):
    keylist = list(d.keys())
    if tlist[t] in keylist:
        print(d[tlist[t]])
    else:
        print(tlist[t])
```
