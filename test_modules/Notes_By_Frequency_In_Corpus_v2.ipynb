# Notes to Frequency:
## 1 piece, Refactored

## About

In this section, we compile a dictionary of 'scale degree' by frequency. How?

### Process

    0) Load any relevant libraries

    1) Set any relevant constants
    
    2) Create an empty dictionary to collect scale degrees.

    3) Access defined corpus
    
    4) Create a list of pieces that make up each element of the collection

    5) For each piece in the list of pieces
        * select piece
        * get the piece's melody line
        * analyze its key
        * transpose key to chosen tonic, same mode
        * get pitches
        * look through pitches, tally pitch occurences in the dictionary

## Helper Functions

### ListElementsOfCorpus( corpus_chosen , element_list )
    create a list composed of the list that makes up each element of the corpus

```python
def listElementsOfCorpus(corpus_chosen,element_list):
    corpus_accessed = mus.corpus.getComposer(corpus_chosen)
    length_corpus_list = len(corpus_accessed)
    i = 0
    while i < length_corpus_list:
        list_of_pieces = mus.corpus.parse(corpus_accessed[i])
        j = len(list_of_pieces)
        for k in range(0,j):
            element_list.append(list_of_pieces[k])
        i+=1
    return(element_list)
    
```

### transposeMelody( melody_line , final_major_key , final_minor_key )
    given a flattened? melody
        1) determine if it is major or minor
        2) output transposed major or minor key

```python
def transposeMelody( melody_line , final_major_key , final_minor_key ):
    """
        - take in flattened? melody
        - determine if melody major or minor key
        - run other helpers transposing key to major or minor
    """
    melody_key = melody_line.analyze('key')
    melody_mode = melody_key.mode
    melody_tonic = melody_key.tonic
    if melody_mode == 'major':
        return(transposeMelodyMajor( melody_line , melody_tonic , final_major_key))
    elif  melody_mode == 'minor':
        return(transposeMelodyMinor( melody_line , melody_tonic , final_minor_key))                 
    else:
        print('Failed to tranpose melody.\n')
```

### transposeMelodyMajor( melody_line , current_major_key ,  final_major_key )
    given a melody in a major key, transpose it to the final key

```python
def transposeMelodyMajor( melody_line, melody_tonic , final_major_key ):
    """
    - have flattened melody line as input
    - figure out how many half steps between tonic and final key
    - transpose stream
    - return transposed stream
    """ 
    current_tonic = mus.note.Note(melody_tonic)
    final_tonic = mus.note.Note(final_major_key)
   
          
    interval_wanted = mus.interval.notesToChromatic( current_tonic , final_tonic )
    transposed_melody = melody_line.flat.transpose(interval_wanted.getDiatonic().directedName)
    
    return(transposed_melody)
```

### transposeMelodyMinor( melody_line , current_major_key , final_minor_key )
